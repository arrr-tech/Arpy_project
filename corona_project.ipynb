```python
for states in coronadata_cases['Location'].tolist():
    Germany_states_list = Germany_map['NAME_1'].tolist()
    if states in Germany_states_list:
        pass
    else:
        print(states + 'state name not found')
```

```python
Germany_map.replace('Nordrhein-Westfalen','North Rhine-Westphalia', inplace = True)
Germany_map.replace('Bayern','Bavaria', inplace = True)
Germany_map.replace('Baden-Württemberg','Baden-WÃ¼rttemberg', inplace = True)
Germany_map.replace('Niedersachsen','Lower Saxony', inplace = True)
Germany_map.replace('Sachsen','Saxony', inplace = True)
Germany_map.replace('Rheinland-Pfalz','Rhineland-Palatinate', inplace = True)
Germany_map.replace('Thüringen','Thuringia', inplace = True)
Germany_map.replace('Sachsen-Anhalt','Saxony-Anhalt', inplace = True)
```
